```python
# This is a machine learning project in which unemployment in India is analysed in various forms to represt the various correlation
# which are not easily identified
# We use a publically available data set from Kaggle: https://www.kaggle.com/datasets/gokulrajkmv/unemployment-in-india
# In this data set we have various unemployment related information
# This project uses various libraries including numpy, pandas, matplotlib.pyplot, seaborn and plotly.express
```

# First we have to import all the necessary libraries which we are goint to use to plot various correations and perform ananlysis

```python
import numpy as np  #linear algebra and supports various other functionalities 
import pandas as pd  #data processing and it has various data structures useful in this project
import seaborn as sns  # creation of statistical graphics and visualizations
import plotly.express as px # we will create two charts using plotly.express i.e. bar and line charts
import matplotlib.pyplot as plt # it is used to create a scatter plot and histogram in this project
```

# Now we will read the dataset using read_csv() function

```python
dataset = pd.read_csv("/content/Unemployment_Rate_upto_11_2020.csv")
print(dataset.head(10)) #printing starting 10 entries from the dataset
print(dataset.tail(10)) #printing ending 10 entrires from the dateset
```

# We can also check the information related to the data set using 'info' method

```python
dataset.info
```

# It is a good practise to check if the data set contains any null, non assigned or empty values

```python
print(dataset.isnull().sum())
```

We can see that the dataset does not have any empty values

# We will now rename the columns in the dataset for better understanding and accessibility

```python
dataset.columns = ["State" , "Date" , "Frequency" , "Estimated Unemployment Rate" , "Estimated Employed" ,
                   "Estimated Labour Participation Rate" , "Region" , "Longitude" , "Latitude"];
```

# Now we can observe the changes made to the dataset

```python
print(dataset)
```

# Now we will display a correlation between the features of the dataset

```python
# creating a heatmap of the correlation matrix
plt.style.use("seaborn-whitegrid") #setting the style of the plot to seaborn-whitegrid
plt.figure(figsize = (12,10)) #figsize is used to determine the width and size of the figure in inches
sns.heatmap(dataset.corr()) #heatmap() creates a heatmap
plt.show()
```

# Displaying a histogram that will show the number of employees in different regions of India. (Employment Statistics)

```python
plt.figure(figsize = (12,10))
plt.title("Employed") # giving title to the histogram
sns.histplot(x = "Estimated Employed" , hue = "Region" , data = dataset) #data stores the dataset , hue is used to set the colorbars
plt.show()
```

# Diplaying unemployment in India based on the number of unemployed people in different regions in India.

```python
plt.figure(figsize = (12,10))
plt.title("Unemployed") # giving title to the histogram
sns.histplot(x = "Estimated Unemployment Rate" , hue = "Region" , data = dataset) #data stores the dataset , hue is used to set the colorbars
plt.show()
```

# Now we will create a Dashboard for the analysis of unemployment in India state-wise. We will use a sunburstplot to plot the unemployment data based on State by region

```python
unemploymentdata = dataset[["State" , "Region" , "Estimated Unemployment Rate"]]
fig = px.sunburst(unemploymentdata , path = ["Region" , "State"]
                  , values = "Estimated Unemployment Rate" ,
                  width = 700 , height = 700 , color_continuous_scale = "RdY1Gn" ,
                  title = "Unemployment Rate in India")
fig.show()
```
